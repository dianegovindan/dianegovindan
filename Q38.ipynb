```python
df_football = pd.read_csv('https://raw.githubusercontent.com/dianegovindan/dianegovindan/main/football_players.csv', encoding='latin-1')
df_football.head()
```

```python
df_football['Overall'].value_counts().head(1)
```

```python
df_cb = df_football[df_football['Preferred Positions'].str.contains('CB')]
df_cb.head()
```

```python
def class_find(score):

    class_cat = 0

    if score >=80:
        class_cat = 1
    elif score<70:
        class_cat = 3
    else:
        class_cat = 2

    return class_cat
```

```python
# create class
df_cb['class'] = df_cb['Overall'].apply(class_find)
df_cb.tail(3000)
```

```python
#target is classes

y = df_cb['class']
```

```python
y.head()
```

```python
X = df_cb.drop(['Preferred Positions','Preferred Positions Type','class','Name','Nationality','Age'], axis = 1)
X.head()
```

```python
#split train-test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.33, random_state = 911)
```

```python
def k_model(k, x_trn, y_trn, x_tst, y_tst):

    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(x_trn, y_trn)
    y_pred = knn.predict(x_tst)

    return f1_score(y_tst, y_pred, average = None) #print(classification_report(y_tst, y_pred))#, target_names=['World Class', 'Good', 'Mediocre']))
```
